```python
import numpy as np 
import matplotlib.pyplot as plt 
#calculus concept 1: the derivative of x² is 2x
#in ML this is exactly what happens during back propogation
#the network asks: if i change this weight slightly, how deos the loss change?

x = np.linspace(-3, 3, 100)
y = x**2      #our function
dy = 2*x      #its derivative - the gradient

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = x²", color="#E94560",linewidth=2 )
plt.plot(x, dy, label= "f'(x) = 2x ← this is the gradient", color="#1A6B9A", linewidth=2, linestyle="--")
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray',linewidth=0.5)
plt.title("Day 1 - Function and its gradient", fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("The gradient at x=2 is:", 2*2)
print("This means: at x=2, if x in the OPPOSITE direction to reduce f(x)")
```

```python
# Gradient descent in 10 lines- this is the heart of all ML
# We want ot find the x that mininises f(x) = x²
# Answer is obviously 0, but let the algorithm find it

x= 10.0  # start far from the minimum
lr=0.1     # learning rate - how big each step is

print("Finding the minimum of f(x) = x² using gradient descent\n")
for step in range(20):
    gradient = 2*x  # derivative of f(x) = x²
    x = x - lr * gradient  # move opposite to the gradient
    print(f"Step {step+1:2d} | x={x:8.4f} | f(x) ={x**2:8.4f}")

    print(f"\nFinal answer: x = {x:.6f} (true minimum is 0)")
    
```
